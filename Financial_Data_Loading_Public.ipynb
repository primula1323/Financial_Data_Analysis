## CCXT

CCXT : CryptoCurrency eXchange Trading Library

JAVAScript/Python 지원. 여러 암호화폐 거래소와 통합되는 라이브러리 및 실시간 데이터 처리 가능.

국내 거래소 중 upbit, bithumb을 지원하며, 국외 거래소에서는 binance 등을 지원.

https://github.com/ccxt/ccxt

```python
#%pip install ccxt --upgrade --no-cache-dir
```

```python
import ccxt
print(ccxt.exchanges) # print a list of all available exchange classes
```

```python
import ccxt

# 거래소 인스턴스 생성 (예: Binance)
exchange = ccxt.binance({
    'apiKey': binance_api_key,
    'secret': binance_secret_key,
})

# BTC/USDT 마켓의 현재 가격 조회
btc_usdt = exchange.fetch_ticker('BTC/USDT')
print(btc_usdt['last'])  # 마지막 거래 가격 출력

# 사용자 계좌 잔액 조회
balance = exchange.fetch_balance()
print(balance['total'])  # 총 잔액 출력
```
